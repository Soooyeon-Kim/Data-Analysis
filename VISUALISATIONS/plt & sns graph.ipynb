## corr
* 각 변수간의(열 간의) 상관관계
> **주의할 점**
1. 연속형(숫자로 표현 가능한) 데이터에 대해서만 상관관계 분석이 가능
2. -1부터 1까지의 값으로 상관관계의 크기를 보여줌
3. 증가하는 방향성에 대한 것. -> 인과관계를 의미하지는 않는다.

```python
titanic.corr()
```

## heatmap

```python
plt.figure(figsize=(10,10))
sns.heatmap(titanic.corr(), annot=True, fmt = '.2f', linewidths=.5, cmap='PuRd')
plt.show()
```

```python
sns.clustermap(titanic.corr(), 
               annot = True,      # 실제 값 화면에 나타내기
               cmap = 'PiYG_r',  # Red, Yellow, Blue 색상으로 표시
               vmin = -1, vmax = 1, #컬러차트 -1 ~ 1 범위로 표시
              )
```

```python
fig, ax = plt.subplots( figsize=(7,7) )

# 삼각형 마스크를 만든다(위 쪽 삼각형에 True, 아래 삼각형에 False)
mask = np.zeros_like(titanic.corr(), dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# 히트맵을 그린다
sns.heatmap(titanic.corr(), 
            cmap = 'RdYlBu_r', 
            annot = True,   # 실제 값을 표시한다
            mask=mask,      # 표시하지 않을 마스크 부분을 지정한다
            linewidths=.5,  # 경계면 실선으로 구분하기
            cbar_kws={"shrink": .5},# 컬러바 크기 절반으로 줄이기
            vmin = -1,vmax = 1   # 컬러바 범위 -1 ~ 1
           )  
plt.show()
```
