## Pokemon vis scatter

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
pkemon = pd.read_csv('./pokemon.csv')
```

```python
# Legendary 포켓몬만 분류하기
legendary = pkemon[pkemon['Legendary']==1]
```

```python
legendary_attack = legendary.sort_values(by='Attack',ascending=True)
legendary_attack
fig, ax = plt.subplots(figsize = (12,6))
plt.scatter(legendary_attack['Name'], legendary_attack['Attack'])
plt.title('Legendary Attack')
plt.xticks(rotation=90)
plt.subplots_adjust(bottom=0.35)

plt.savefig("graph.png")
```
