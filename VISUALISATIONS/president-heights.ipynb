```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("./president_heights.csv")

fig, ax = plt.subplots()
ax.plot(df['order'],
       df['height(cm)'], label='height')

ax.set_xlabel('order')
ax.set_ylabel('height(cm)')
```

```python
dfh = pd.DataFrame(df)
dfh.head()
```

```python
heights = np.array(dfh['height(cm)'])
heights
```

```python
print("mean height :",heights.mean())
print("standard deviation :",heights.std())
print("min height :",heights.min())
print("max height :",heights.max())
```

```python
print("25th percentile:", np.percentile(heights, 25)) # 25번째 사분위수
print("median:", np.median(heights))
print("75th percentile:", np.percentile(heights, 75))# 75번째 사분위수
```

```python
plt.hist(heights)
plt.title("Heignt dist of US presidents")
plt.xlabel('height(cm)')
plt.ylabel('order')
```
