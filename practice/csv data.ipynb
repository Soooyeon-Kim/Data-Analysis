```python
import csv
```

`reader()`는 CSV 파일의 내용을 먼저 줄 별로 나눈 뒤, 구분 기호(`delimiter`)를 기준으로 분리

books.csv 파일은 1열부터 제목, 저자, 장르, 페이지 수, 출판사 순서로 나열

```python
def print_book(filename):
    with open(filename) as file:    
        reader = csv.reader(file, delimiter=',')
        for row in reader:
            print(row)

            
file = 'books.csv'
print_book(file)
```

```python
def print_book(filename):
    with open(filename) as file:    
        reader = csv.reader(file, delimiter=',')
        for row in reader:
            print(row[0])

            
file = 'books.csv'
print_book(file)
```

제목 (저자): 페이지 수p 형식

```python
def print_book_info(filename):
    with open(filename) as file:
        reader = csv.reader(file, delimiter=',')
        for row in reader:
            title = row[0]
            author = row[1]
            pages = row[3]
            print("{} ({}): {}p".format(title, author, pages))

filename = 'books.csv'
print_book_info(filename)
```
